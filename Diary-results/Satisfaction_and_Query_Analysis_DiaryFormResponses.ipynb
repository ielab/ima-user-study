```python
# for i in si:
#   for index, row in df1["What was your search intent when issuing this query to search your historical chat?"].items():
#     if i == row:
#       df3.at[index, i] = df1.loc[index, "How satisfied were you with the results being displayed in that way?"]




result = df.groupby(df["What chat objects satisfied your search intents when you used the search function?"]).agg({"What chat objects satisfied your search intents when you used the search function?": 'first', 'How satisfied were you with the quality of the search results?': 'mean'})

# for index, row in df["How satisfied were you with the results being displayed in that way?"].items():
#   print(row)

result2 = df.groupby("What chat objects satisfied your search intents when you used the search function?")["How satisfied were you with the quality of the search results?"].count()
result["vote_count"] = result2
result.head(100)
```

```python
# for i in si:
#   for index, row in df1["What was your search intent when issuing this query to search your historical chat?"].items():
#     if i == row:
#       df3.at[index, i] = df1.loc[index, "How satisfied were you with the results being displayed in that way?"]




result_2 = df.groupby(df["What chat objects satisfied your search intents when you used the search function?"]).agg({"What chat objects satisfied your search intents when you used the search function?": 'first', 'How satisfied were you with the results being displayed in that way?': 'mean'})

# for index, row in df["How satisfied were you with the results being displayed in that way?"].items():
#   print(row)

result_3 = df.groupby("What chat objects satisfied your search intents when you used the search function?")["How satisfied were you with the results being displayed in that way?"].count()
result_2["vote_count_2"] = result_3
result_2.head(100)
```

```python
C = result['How satisfied were you with the quality of the search results?'].mean()
C2 = result_2['How satisfied were you with the results being displayed in that way?'].mean()

print(C)
print(C2)
```

```python
m = result["vote_count"].quantile(0.9)
m2 = result_2["vote_count_2"].quantile(0.9)

print(m)
print(m2)
```

```python
# q_intents = result.copy().loc[result['vote_count'] >= m]
# print(q_intents.shape)
```

```python
def weighted_rating(x, m=m, C=C):
   v = x['vote_count']
   R = x['How satisfied were you with the quality of the search results?']
   return (v/(v+m) * R) + (m/(m+v) * C)


# q_intents['score'] = q_intents.apply(weighted_rating, axis=1)
result['weighted_rating'] = result.apply(weighted_rating, axis=1)
```

```python
result_2.head(100)
```

```python
def weighted_rating2(x, m=m2, C=C2):
   v = x['vote_count_2']
   R = x['How satisfied were you with the results being displayed in that way?']
   return (v/(v+m) * R) + (m/(m+v) * C)

result_2['weighted_rating2'] = result_2.apply(weighted_rating2, axis=1)
```

```python
result_2.head(100)
```

```python
final_result = pd.merge(result,result_2)
final_result.head(1900)
```
